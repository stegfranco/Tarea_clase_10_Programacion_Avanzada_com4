# Patrones de Diseño en Programación

Los **patrones de diseño** son soluciones reutilizables a problemas comunes en el desarrollo de software. Se originaron en la ingeniería de software orientada a objetos, pero sus principios son aplicables a distintos lenguajes y paradigmas.

## Categorías de Patrones de Diseño

Los patrones de diseño se agrupan en **tres grandes categorías**:

1. **Patrones Creacionales**: Se enfocan en el proceso de creación de objetos. Permiten crear objetos de manera controlada, flexible y desacoplada.
2. **Patrones Estructurales**: Se ocupan de la composición de clases y objetos para formar estructuras más complejas, facilitando su uso conjunto.
3. **Patrones de Comportamiento**: Se centran en la comunicación entre objetos y cómo interactúan entre sí.

A continuación, se presentan ejemplos representativos de cada categoría con explicaciones y contexto de uso.


## Patrón Singleton (Creacional)

El patrón **Singleton** asegura que una clase tenga **una única instancia** y proporciona un punto de acceso global a ella.

### Contexto
Se utiliza cuando se necesita exactamente **una instancia de una clase** para coordinar acciones en todo el sistema, como un administrador de configuración o una conexión a base de datos.

### Ventajas
- Control del acceso a la instancia.
- Reducción del uso de recursos.

### Explicación paso a paso del patrón Singleton con un ejemplo
1. Se define una clase `Singleton` que tiene un atributo de clase `_instance` inicializado como `None`.
2. Se sobrescribe el método especial `__new__`, que controla la creación de instancias.
3. Si `_instance` es `None`, se crea una nueva instancia de `Singleton`.
4. En llamadas posteriores, se devuelve la instancia ya existente.
5. En el ejemplo, `s1` y `s2` son exactamente el mismo objeto, por lo que `s1 is s2` es `True`.


```python
class Singleton:
    _instance = None

    def __new__(cls):
        if cls._instance is None:
            cls._instance = super(Singleton, cls).__new__(cls)
            print("Instancia creada")
        return cls._instance

# Ejemplo de uso
s1 = Singleton() # Instancia creada
s2 = Singleton()
print(s1 is s2)  # True - ambas son la misma instancia
```

## Patrón Factory Method (Creacional)

El patrón **Factory Method** define una **interfaz para crear objetos**, pero permite a las subclases decidir qué clase concreta instanciar. Promueve el principio de **abierto/cerrado**, ya que se pueden añadir nuevos productos sin modificar el código existente.

### Contexto
Se usa cuando un código no debe depender de cómo se crean o configuran los objetos. Por ejemplo, una aplicación puede requerir diferentes interfaces gráficas para distintos sistemas operativos.

### Ventajas
- Desacopla la lógica de creación de objetos.
- Facilita la extensibilidad.

### Explicación paso a paso del patrón Factory Method con un ejemplo
1. Se define una interfaz `Product` con un método común (`operation()`).
2. Se crean implementaciones concretas de `Product`, como `ConcreteProductA` y `ConcreteProductB`.
3. La clase `Creator` declara el método `factory_method()` (abstracto) y un método común `some_operation()` que usa productos.
4. Las subclases `ConcreteCreatorA` y `ConcreteCreatorB` implementan el `factory_method()` y devuelven instancias del producto adecuado.
5. En el uso, el cliente interactúa con el creador abstracto y obtiene objetos sin conocer su clase concreta.


```python
class Product:
    def operation(self):
        pass

class ConcreteProductA(Product):
    def operation(self):
        return "Resultado del Producto A"

class ConcreteProductB(Product):
    def operation(self):
        return "Resultado del Producto B"

class Creator:
    def factory_method(self):
        pass

    def some_operation(self):
        product = self.factory_method()
        return product.operation()

class ConcreteCreatorA(Creator):
    def factory_method(self):
        return ConcreteProductA()

class ConcreteCreatorB(Creator):
    def factory_method(self):
        return ConcreteProductB()

# Ejemplo de uso
creator = ConcreteCreatorA()
print(creator.some_operation())
```

## Patrón Adapter (Estructural)

El patrón **Adapter** permite que **interfaces incompatibles colaboren** mediante una clase que traduce una interfaz en otra que el cliente espera.

### Contexto
Se usa cuando se quiere integrar una clase existente en una aplicación sin modificar su código original, pero su interfaz no es compatible con lo que se espera.

### Ventajas
- Reutilización de código existente.
- Flexibilidad para integrar componentes externos.

### Explicación paso a paso del patrón Adapter con un ejemplo
1. La clase `OldPrinter` representa una impresora antigua con una interfaz obsoleta (`old_print()`).
2. El código cliente espera una interfaz más moderna, con un método llamado `print()`.
3. En lugar de modificar `OldPrinter`, se crea una clase `PrinterAdapter` que **traduce** el método `print()` al método `old_print()` del objeto original.
4. En el uso práctico, se crea un objeto `OldPrinter`, se pasa al `PrinterAdapter`, y se llama al método `print()` sin saber que internamente está usando `old_print()`.


```python
# Interfaz Impresora
class Printer:
    def print(self):
      pass

class OldPrinter:
    def old_print(self):
        return "Impresión vieja"

class PrinterAdapter(Printer):
    def __init__(self, old_printer):
        self.old_printer = old_printer

    def print(self):
        return self.old_printer.old_print()

# Ejemplo de uso
old = OldPrinter()
adapter = PrinterAdapter(old)
print(adapter.print())
```

## Patrón Decorator (Estructural)

El patrón **Decorator** permite añadir funcionalidades a objetos **de forma dinámica** sin modificar su estructura original.

### Contexto
Es útil cuando se desea añadir responsabilidades adicionales a un objeto sin alterar su código, especialmente cuando no es posible usar la herencia.

### Ventajas
- Aumenta la flexibilidad al extender funcionalidades.
- Permite componer comportamientos en tiempo de ejecución.

### Explicación paso a paso del patrón Decorator con un ejemplo
1. Se define la clase base `Notifier` con un método `send()` que será decorado.
2. `BaseDecorator` acepta un objeto `Notifier` y delega la llamada a `send()` para mantener el comportamiento original.
3. `SMSDecorator` extiende `BaseDecorator` y añade funcionalidad adicional (`vía SMS`) sin modificar `Notifier`.
4. Finalmente, se crea un objeto `decorated` que añade la funcionalidad extra sobre la instancia base.


```python
class Notifier:
    def send(self):
        return "Enviando notificación"

class BaseDecorator(Notifier):
    def __init__(self, notifier):
        self._notifier = notifier

    def send(self):
        return self._notifier.send()

class SMSDecorator(BaseDecorator):
    def send(self):
        return f"{super().send()} vía SMS"

# Ejemplo de uso
notifier = Notifier()
decorated = SMSDecorator(notifier)
print(decorated.send())
```

## Patrón Observer (Comportamiento)

El patrón **Observer** permite que múltiples objetos (observadores) escuchen y reaccionen a los **cambios de estado** de otro objeto (sujeto).

### Contexto
Es ideal cuando un cambio en un objeto requiere actualizaciones en otros objetos, y no se sabe cuántos objetos necesitan ser actualizados.

### Ventajas
- Promueve el bajo acoplamiento.
- Facilita la comunicación uno-a-muchos.

### Explicación paso a paso del patrón Observer con un ejemplo
1. La clase `Subject` gestiona una lista de observadores (`_observers`).
2. Los observadores se registran mediante el método `attach()`.
3. Cuando hay una actualización, se llama a `notify()`, que ejecuta `update()` en cada observador.
4. La clase `ConcreteObserver` implementa `update()` para reaccionar al evento.


```python
class Subject:
    def __init__(self):
        self._observers = []

    def attach(self, observer):
        self._observers.append(observer)

    def notify(self, message):
        for observer in self._observers:
            observer.update(message)

class Observer:
    def update(self, message):
        pass

class ConcreteObserver(Observer):
    def update(self, message):
        print(f"Notificado con mensaje: {message}")

# Ejemplo de uso
subject = Subject()
observer = ConcreteObserver()
subject.attach(observer)
subject.notify("Actualización importante")
```

## Patrón Strategy (Comportamiento)

El patrón **Strategy** permite definir una familia de algoritmos y hacerlos **intercambiables**. El algoritmo se selecciona en tiempo de ejecución.

### Contexto
Se utiliza cuando una clase debe poder usar diferentes variantes de un algoritmo sin modificar su estructura.

### Ventajas
- Facilita el mantenimiento y la extensión de algoritmos.
- Cumple con el principio de abierto/cerrado.

### Explicación paso a paso del patrón Strategy con un ejemplo

1. Se define una interfaz `Strategy` con el método `execute()`.
2. `AddStrategy` y `SubtractStrategy` implementan esta interfaz con distintos comportamientos.
3. La clase `Context` almacena una instancia de una estrategia y la usa cuando se llama `execute_strategy()`.
4. En el uso, se cambia la estrategia dinámicamente para aplicar suma o resta según se necesite.



```python
class Strategy:
    def execute(self, a, b):
        pass

class AddStrategy(Strategy):
    def execute(self, a, b):
        return a + b

class SubtractStrategy(Strategy):
    def execute(self, a, b):
        return a - b

class Context:
    def __init__(self, strategy):
        self.strategy = strategy

    def execute_strategy(self, a, b):
        return self.strategy.execute(a, b)

# Ejemplo de uso
context = Context(AddStrategy())
print(context.execute_strategy(5, 3))  # 8
context.strategy = SubtractStrategy()
print(context.execute_strategy(5, 3))  # 2
```
